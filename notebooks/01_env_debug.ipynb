```python
# notebooks/01_env_debug.ipynb

from envs.battalion_env import BattalionEnv
from stable_baselines3.common.env_checker import check_env
import numpy as np

env = BattalionEnv()

# 1. SB3's built-in sanity checker — catches most bugs
check_env(env, warn=True)
print("✅ Environment passed check_env")

# 2. Random rollout — make sure nothing crashes
obs, info = env.reset(seed=42)
print(f"Observation shape: {obs.shape}")
print(f"Sample obs: {obs}")

total_reward = 0
for step in range(200):
    action = env.action_space.sample()   # random actions
    obs, reward, terminated, truncated, info = env.step(action)
    total_reward += reward
    if terminated or truncated:
        print(f"Episode ended at step {step}, total reward: {total_reward:.2f}")
        obs, info = env.reset()
        total_reward = 0

print("✅ Random rollout completed without errors")

# 3. Sanity check reward range
rewards = []
obs, _ = env.reset()
for _ in range(1000):
    action = env.action_space.sample()
    obs, r, terminated, truncated, _ = env.step(action)
    rewards.append(r)
    if terminated or truncated:
        obs, _ = env.reset()

print(f"Reward stats — min: {min(rewards):.2f}, max: {max(rewards):.2f}, "
      f"mean: {np.mean(rewards):.2f}")
```
